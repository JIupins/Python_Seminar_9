```python
# 1 Импорт библиотеки для работы с табличными данными
import pandas as pd
import seaborn as sb
```

```python
# 2 Чтение данных из файла, с помощю метода библиотеки Pandas
df = pd.read_csv('sample_data/california_housing_train.csv')
```

```python
# 3 Вывод данных из файла
df
```

```python
# 4 Дома где кол-во людей от 0 до 500 (population)
dv = df.loc[     df['population'] <= 500,   ['median_house_value']    ]
```

```python
# 5 Средняя стоимость дома где кол-во людей от 0 до 500 (population)
dv.mean()
```
